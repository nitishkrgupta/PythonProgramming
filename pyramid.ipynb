```python
n = 5
spc = n - 1
star = 1

for i in range(n):
    for j in range(spc):
        print(" ", end=" ")
    for k in range(star):
        print("*", end=" ")
    print()
    spc -= 1
    star += 2
```

```python
n = 5

for i in range(n):
    for j in range(n - 1 - i):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*", end=" ")
    print()
```

```python
n = 4
spc = 0
str = n * 2 - 1

for i in range(n):
    for j in range(spc):
        print(" ", end=" ")
    for k in range(str):
        print("*", end=" ")
    print()
    spc += 1
    str -= 2
```
