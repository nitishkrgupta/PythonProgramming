```python
n = int(input("Enter n : "))

for i in range(n):
    for j in range(n+2):
        if i == n//2 or i + j == 2*n-2 or j - i == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
import math as m
# pascal triangle

n = int(input("Enter n : "))

for i in range(n):
    for j in range(n-1-i):
        print(" ", end="")
    for k in range(i+1):
        print(m.comb(i, k), end=" ")
    print()
```
