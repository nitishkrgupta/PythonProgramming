```python
# printing alphabates in square or rectangle shape
row = int(input("Enter no. of row : "))
col = int(input("Enter no. of col : "))
val = ord('A')
for i in range(row):
    for j in range(col):
        print(chr(val), end=" ")
    print()
    val += 1
    if val > ord('Z'): val = ord('A')
```

```python
n = int(input("Enter n : "))
val = 1
p = True
for i in range(n):
    for j in range(n):
        if p:
            print(val, end=" ")
            val += 1
            if val > 9: val = 1
            p = False
        else:
            print("*", end=" ")
            p = True
    print()
```

### Triangular pattern

```python
n = int(input("n : "))

# for i in range(n):
#     for j in range(i):
#         print("*", end=" ")
#     print()

for i in range(n):
    for j in range(n):
        if i >= j:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n = int(input("Enter n : "))
for i in range(n):
    for j in range(n):
        if i <= j:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n = int(input("Enter n : "))
for i in range(n):
    for j in range(n):
        if i == j:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n = int(input("Enter n : "))
for i in range(n):
    for j in range(n):
        if i + j == (n - 1):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
